# Iteration 

## 1D

```python
import numpy as np
var = np.array([9, 8, 7, 6, 5, 4])
print(var)
print()

for i in var:
    print(i)
```

## 2D

```python
var1 = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])
print(var1)
print()

for j in var1:
    print(j)

print()

for k in var1:
    for l in k: 
        print(l)
```

## 3D

```python
var3 = np.array([[[1, 2, 3, 4], [1, 2, 3, 4]]])
print(var3)
print(var3.ndim)
print()

for i in var3:
    for j in i:
        for k in j:
            print(k)
```

### np.nditer() function

```python
var3 = np.array([[[9, 8, 7, 6], [1, 2, 3, 4]]])
print(var3)
print(var3.ndim)
print()

for i in np.nditer(var3):
    print(i)
```

```python
var3 = np.array([[[1, 2, 3, 4], [1, 2, 3, 4]]])
print(var3)
print(var3.ndim)
print()

for i in np.nditer(var3, flags = ['buffered'], op_dtypes = ["S"]):
    print(i)
```

### np.ndenumerate() function

```python
var3 = np.array([[[9, 8, 7, 6], [1, 2, 3, 4]]])
print(var3)
print(var3.ndim)
print()

for i, d in np.ndenumerate(var3):
    print(i, d)
```
