# Matrix

## Array vs Matrix

```python
import numpy as np
var = np.matrix([[1, 2, 3], [1, 2, 3]])
print(var)
print(type(var))
```

```python
var1 = np.array([[1, 2, 3], [1, 2, 3]])
print(var1)
print(type(var1))
```

## Arithmetic operation in matrix

### Addition 

```python
var1 = np.matrix([[1, 2, 3], [1, 2, 3]])
print(var1)
print()
var2 = np.matrix([[1, 2, 3], [1, 2, 3]])
print(var1 + var2)
```

### Subtraction

```python
var1 = np.matrix([[1, 2, 3], [1, 2, 3]])
print(var1)
print()
var2 = np.matrix([[1, 2, 3], [1, 2, 3]])
print(var1 - var2)
```

### Multiplication

```python
var1 = np.matrix([[1, 2], [1, 2]])
print(var1)
print()
var2 = np.matrix([[1, 2], [1, 2]])
print(var1 * var2)
print()

print(var1.dot(var2))
```
