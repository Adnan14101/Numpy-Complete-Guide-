## Arithmetic Functions

### Min and Max

```python
import numpy as np

var = np.array([1, 2, 3, 4, 5, 3, 2])

print("Min : ", np.min(var))
print("Max : ", np.max(var))
```

### Position

```python
var = np.array([1, 2, 3, 4, 5, 3, 2])

print("Min : ", np.min(var))
print("Min Position :", np.argmin(var))

print("Max : ", np.max(var))
print("Max Position : ", np.argmax(var))
```

## 2D Array

```python
var1 = np.array([[2, 1, 3], [9, 5, 6]])

print(np.min(var1, axis = 0))  # axis = 0 for column
```

```python
var1 = np.array([[2, 1, 3], [9, 5, 6]])

print(np.min(var1, axis = 1))  # axis = 1 for Row
```

### Square root (sqrt)

```python
import numpy as np

var = np.array([1, 2, 3, 4, 5, 3, 2])

print("Min : ", np.min(var))
print("Max : ", np.max(var))
print("sqrt : ", np.sqrt(var))
```

### Sin and Cos 

```python
var2 = np.array([1, 2, 3])

print(np.sin(var2))
print(np.cos(var2))
```

### Cummulative Sum (cumsum)

```python
var3 = np.array([1, 2, 3])

print(np.cumsum(var3))
```
