# Matrix Functions

## Transpose

```python
import numpy as np
var = np.matrix([[1, 2, 3], [4, 5, 6]])
print(var)
print()

print(np.transpose(var))
print()

print(var.T)
```

## Swapaxes

```python
var1 = np.matrix([[1, 2, 3], [4, 5, 6]])
print(var1)
print()

print(np.swapaxes(var1, 0, 1))
```

```python
var2 = np.matrix([[1, 2], [4, 5]])
print(var2)
print()

print(np.swapaxes(var2, 0, 1))
```

## Inverse

```python
var3 = np.matrix([[1, 2], [3, 4]])
print(var3)
print()

print(np.linalg.inv(var3))
```

## Power

```python
var4 = np.matrix([[1, 2], [3, 4]])
print(var4)
print()

print(np.linalg.matrix_power(var4, 2))
print()

print(np.linalg.matrix_power(var4, 0))
print()

print(np.linalg.matrix_power(var4, -2))
print()
```

## Determinate

```python
var5 = np.matrix([[1, 2], [3, 4]])
print(var5)
print()

print(np.linalg.det(var5))
```

```python
var6 = np.matrix([[1, 2, 3], [3, 4, 3], [1, 2, 3]])
print(var6)
print()

print(np.linalg.det(var6))
```
