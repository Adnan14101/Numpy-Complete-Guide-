## Arithmetic Operations

### Add

```python
import numpy as np
var = np.array([1, 2, 3, 4])
varadd = var + 3
print(varadd)
```

```python
var1 = np.array([1, 2, 3, 4])
var2 = np.array([1, 2, 3, 4])

varadd = var1 + var2

print(varadd)
```

```python
var1 = np.array([1, 2, 3, 4])
var2 = np.array([1, 2, 3, 4])

varadd = np.add(var1, var2)

print(varadd)
```

```python
var1 = np.array([1, 2, 3, 4])
var2 = np.array([1, 2, 3, 4])

varadd = np.add(var1, 3)

print(varadd)
```

#### 2D array

```python
var2 = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])
var3 = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])

print(var2)
print("+")
print(var3)
print("=")

varadd2 = var2 + var3
print(varadd2)
```

### Subtraction

### Power

```python
var = np.array([1, 2, 3, 4])
varadd = var ** 3
print(varadd)
```

#### 2D Array

```python
var2 = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])
var3 = np.array([[1, 2, 3, 4], [1, 2, 3, 4]])

print(var2)
print("**")
print(var3)
print("=")

varadd2 = var2 ** var3
print(varadd2)
```

### Reciprocal

```python
var = np.array([1, 2, 3, 4])
varadd = np.reciprocal(var)
print(varadd)
```
