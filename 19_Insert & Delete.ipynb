# Insert and Delete

## Insert

```python
import numpy as np
var = np.array([1, 2, 3 ,4])
print(var)
v = np.insert (var, (2, 4), 40)
print(v)
```

#### Append function

```python
var = np.array([1, 2, 3 ,4])
print(var)
x = np.append (var, 6.5)
print(x)
```

### 2D

```python
var1 = np.array([[1, 2, 3], [1, 2, 3]])

v1 = np.insert(var1, 2, [25, 26], axis = 1)

print(v1)
```

```python
var1 = np.array([[1, 2, 3], [1, 2, 3]])

v1 = np.insert(var1, 2, [23, 24, 25], axis = 0)

print(v1)
```

#### Append function

```python
var1 = np.array([[1, 2, 3], [1, 2, 3]])

v1 = np.append(var1, [[23, 24, 25]], axis = 0)

print(v1)
```

## Delete function

```python
var2 = np.array([1, 2, 3 ,4])
print(var2)
d = np.delete(var2, 2)
print(d)
```
