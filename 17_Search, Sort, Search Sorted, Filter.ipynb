# Search, Search Sorted, Sort, Filter Array

## Search Array :- Search an array for a certain value and return the indexes that get a match

```python
import numpy as np
var = np.array([1, 2, 3, 4, 2, 5, 2, 5, 6, 7])

x = np.where(var == 2)
print(x)
```

```python
var = np.array([4, 2, 3, 4, 2, 5, 2, 5, 6, 7])

x = np.where((var % 2) == 0)
print(x)
```

## Search Sorted Array :- it performs a binary search in the array, and returns the index where the specified value would be inserted to maintain the search order.

```python
var1 = np.array([1, 2, 3, 4, 6, 7, 8])

x1 = np.searchsorted(var1, 5)
print(x1)
```

```python
var1 = np.array([1, 2, 3, 4, 9, 10])

x1 = np.searchsorted(var1, [5, 6, 7], side = "right")
print(x1)
```

## Sort Array :- Ordered sequence is any sequence that has an order corresponding to elements, like numeric or alphabetical, ascending or descending.

```python
var_1 = np.array([4, 2, 3, 1, 12, 15, 22, 5, 52, 6, 7])

print(np.sort(var_1))
```

```python
var_1 = np.array([[4, 2, 3], [1, 12, 15], [22, 5, 52]])

print(np.sort(var_1))
```

```python
var_1 = np.array(["a", "s", "d", "f"])

print(np.sort(var_1))
```

## Filter Array :- Getting some elements out of an existing array and creating a new array out of them.

```python
var_2 = np.array(["a", "s", "d", "f"])

f = [True, False, False, True]
new_a = var_2[f]
print(new_a)
print(type(new_a))
```
