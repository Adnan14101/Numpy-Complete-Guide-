# Zeros

```python
import numpy as np

ar_zero = np.zeros(4)
ar_zero1 = np.zeros((3, 4))

print(ar_zero)
print(ar_zero.ndim)
print()
print(ar_zero1)
print(ar_zero1.ndim)
```

# Ones

```python
import numpy as np

ar_one = np.ones(4)
ar_one1 = np.ones((3,4))

print(ar_one)
print(ar_one.ndim)
print()
print(ar_one1)
```

# Empty

```python
import numpy as np
ar_em = np.empty(4)
print(ar_em)
```

# Range

```python
import numpy as np
ar_rn = np.arange(4)
print(ar_rn)
```

# Diagonal

```python
ar_dia = np.eye(3)
print(ar_dia)
```

# Linspace

```python
ar_lin = np.linspace(0, 20, num = 5)
print(ar_lin)
```
