# Join Array

## 1D

```python
import numpy as np
var = np.array([1, 2, 3, 4])
var1 = np.array([9, 8, 7, 6])

ar = np.concatenate((var, var1))
print(ar)
```

## 2D

```python
vr = np.array([[1, 2], [3, 4]])
vr1 = np.array([[9, 8], [7, 6]])
print(vr)
print()
print(vr1)
print()

ar_new = np.concatenate((vr, vr1), axis = 0)
print(ar_new)

print()

ar_new1 = np.concatenate((vr, vr1), axis = 1)
print(ar_new1)
```

## np.stack() function

```python
var_1 = np.array([1, 2, 3, 4])
var_2 = np.array([9, 8, 7, 6])

a_new = np.stack((var_1,var_2), axis = 1)
print(a_new)
print()

a_new1 = np.hstack((var_1,var_2)) #row
print(a_new1)
print()

a_new2 = np.vstack((var_1,var_2)) #column
print(a_new2)
print()

a_new3 = np.dstack((var_1,var_2)) #height
print(a_new3)
print()
```

# Split Array

## 1D

```python
var = np.array([1, 2, 3, 4, 5, 6])
print(var)
print()

ar = np.array_split(var,3)
print(ar)
print(type(ar))
print(ar[0])
```

## 2D

```python
var1 = np.array([[1, 2], [3, 4], [5, 6]])
print(var1)
print()

ar1 = np.array_split(var1,3)
print(ar1)
print()

ar2 = np.array_split(var1, 3, axis = 1)
print(ar2)
```
